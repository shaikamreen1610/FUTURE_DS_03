## STEP 2 : Rating Analysis (1â€“5 Scale)

```python
weight_cols = ['weightage_1','weightage_2','weightage_3','weightage_4','weightage_5']

df['Total_Responses'] = df[weight_cols].sum(axis=1)

df['Avg_Rating'] = (
    (df['weightage_1']*1 +
     df['weightage_2']*2 +
     df['weightage_3']*3 +
     df['weightage_4']*4 +
     df['weightage_5']*5) / df['Total_Responses']
)

df[['course_name','Avg_Rating']].head()
```

## Average Satisfaction by Course:

```python
def satisfaction(x):
    if x >= 4:
        return "Highly Satisfied"
    elif x >= 3:
        return "Satisfied"
    else:
        return "Not Satisfied"

df['Satisfaction_Level'] = df['Avg_Rating'].apply(satisfaction)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='course_name', y='Avg_Rating', data=df, color = 'green', saturation=0.5, alpha = 0.8, errorbar = 'ci')
plt.xticks(rotation=90)
plt.title("Average Satisfaction Rating by Course")
plt.show()
```

## Distribution of Question Sentiment Categories

```python
sentiment_counts = df['questions'].value_counts().reset_index()
sentiment_counts.columns = ['question', 'count']

plt.figure(figsize=(10,10))
sns.barplot(x='count', y='question', data=sentiment_counts, color = 'Green', alpha = 0.5)
plt.title("Distribution of Question Sentiment Categories")
plt.xlabel("Count")
plt.ylabel("Question")
plt.show()
```

## Sentiment Analysis Distribution

```python
df['Sentiment'] = pd.cut(
    df['Avg_Rating'],
    bins=[0, 3, 4, 5],
    labels=['Negative', 'Neutral', 'Positive'],
    include_lowest=True
)
```

```python
sentiment_counts = df['Sentiment'].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(sentiment_counts.index, sentiment_counts.values , color = 'Green',alpha =0.5)
plt.title("Sentiment Analysis Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Records")
plt.show()
```
